```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from random import randint
# home articles contact admin

interaction = {
    "timestamp": datetime.utcnow,
    "page": "home",
    "buttonID": "upvote"
}

# Take a list of interactions as dictionaries and display a line graph of the data over time. 
def hourly_activity(date):
    """
    Display a line graph of user activity each hour over one day. 
    """
    
    # Create a list with the X coordinates of each bar's left edge
    x_ticks = []
    for i in range(0,24):
        x_ticks.append(i)
    # print(left_edges)
    # Create a list with the heights of each bar - Y coordinates
    y_ticks = [10, 20, 30, 40, 50]
    
    x_coords = list(x_ticks)
    
    
    y_coords = []
    
    for i in range(0,24):
        y_coords.append(randint(0,50))
    
    # Build the line graph and set dot as marker
    plt.plot(x_coords, y_coords, marker = "o")
    
    # Add a title
    plt.title("Hourly Activity for " + date)
    
    # Add a label to the x-axis
    plt.xlabel("Hour")
    
    # Add a label to the y-axis
    plt.ylabel("Number of clicks")

    # Add a grid
    plt.grid(True)
    
    # Display the line graph
    plt.show()
hourly_activity("Jan 25 2022")

# Take a list of interactions as dictionaries and display a bar graph of the data over time. 
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon',
        width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
# Take a list of interactions as dictionaries and display a pie chart of the data over time. 
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```
